## Ejercicio 1
En una farmacéutica, se quiere determinar si el uso de cocaína mejora el resfriado común. 
Se dividen aleatoriamente a 180 personas resfriadas en dos grupos de 90 individuos cada uno. Al grupo 1 se le suministra un placebo, y al grupo 2 la droga. Al día siguiente, 30 de los 90 individuos del grupo 1 aseguran haber mejorado del resfriado. En el grupo 2, 41 de los 90 individuos aseguran haber mejorado. 

¿Se puede concluir que el fármaco experimental tiende a mejorar los síntomas del resfriado?



$p_1 = 30/90$  
$p_2 = 41/90$  

$H_0: p_1 = p_2$    
$H_1: p_1 < p_2$

````python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

exitos = np.array([30, 41])
casos = np.array([90, 90])

(z, pvalor) = proportions_ztest(count=exitos, nobs=casos, alternative='smaller')
print(z)
print(pvalor)
````

## Ejercicio 3
Se realiza un experimento para determinar si el bootcamp en Data Science de The Bridge es mejor que el de Ironhack. Al terminar ambos cursos, se escogen aleatoreamente a alumnos de cada escuela, y se les realiza el mismo examen, obteniendo los siguientes resultados:

Notas The Bridge: 9, 7, 5, 8, 9, 4, 10, 7, 9, 2, 10, 3  
Notas Ironhack: 5, 6, 4, 1, 10, 10, 8, 7, 6, 5

¿Qué escuela es mejor?



````python
bridge = [9, 7, 5, 8, 9, 4, 10, 7, 9, 2, 10, 3]
iron =  [5, 6, 4, 1, 10, 10, 8, 7, 6, 5]

print('The Bridge\n')
print('Media:', np.mean(bridge))
print('Muestra:',len(bridge))

print('\n\nIronhack\n')
print('Media:', np.mean(iron))
print('Muestra:',len(iron))
````

Definimos las hipótesis nula y alternativa

$H_0: \mu_1 = \mu_2$  
$H_1: \mu_1 > \mu_2$

````python
import scipy.stats as stats
stats.ttest_ind(a=bridge, b=iron, alternative='greater')
````
