## 2. Missing data

`NaN` significa *Not a Number*, y en general los reemplazamos por otros valores. 

Podemos observar que:
- `director` tiene el mayor porcentaje de nulos (~30%)
- `cast` y `country` también tienen un porcentaje considerable de nulos (~9%)
- `date_added`, `rating` y `duration` no tienen muchos valores nulos (entre 0% y 0.1%)
- Afortunadamente, muchas columnas están completas.  

¿Qué hacemos con las columnas con nulos? Tenemos dos opciones:
1. Elimina la columna por completo si no es importante para el análisis
2. Mantener la columna y reemplazar los valores nulos. 

Hay que tener cuidado con eliminar los nulos, ya que podemos perder filas que tengan otros campos relevantes

Vemos que la mayoría de las columnas son de tipo `object`, lo que significa que son de tipo string o mixto (valores numéricos y no numéricos mezclados).  
Para las columnas de tipo string, vamos a sustituir los valores `NaN` con "", usando el método `fillna`
