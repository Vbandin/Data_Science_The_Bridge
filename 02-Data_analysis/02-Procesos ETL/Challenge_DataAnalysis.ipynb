```python
url = df.loc[43,'link']
```

```python
import urllib.request
f = urllib.request.urlopen(url)
html = f.read().decode()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html)

tags = soup('span',class_='auto-style21')
book = ''
for tag in tags:
    if len(tag.attrs) == 1:
        book+=tag.text

book = book.replace(',','')\
                   .replace('.','')\
                   .replace(':','')\
                   .replace('?','')\
                   .replace('"','')\
                   .replace('<span class="auto-style21">','')\
                   .replace('</span>','')\
                   .replace('\xa0','')\
                   .replace('”“','')\
                   .replace('\n[','')\
                   .replace('\n','')\
                   .replace(']','')\
                   .replace(' [Continue to [Back to ','')

book
```

```python
contadores2 = {}

palabras = book.lower().split()
palabras
```

```python
for palabra in palabras:
    if len(palabra)==6:
        contadores2[palabra] = contadores2.get(palabra,0) + 1
```

```python
df = pd.DataFrame(list(contadores2.items()), columns = ['palabra','contador'])
df.sort_values('contador',ascending=False)

```
