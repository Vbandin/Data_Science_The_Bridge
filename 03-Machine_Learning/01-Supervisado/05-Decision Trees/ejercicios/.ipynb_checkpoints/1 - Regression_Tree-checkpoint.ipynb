# Ejercicio Regression Tree
Son datos de temperaturas tomados de la ciudad de Seattle. El objetivo es intentar predecir lo mejor posible la máxima temperatura (columna `actual`) que alcanzaremos al día siguiente. Para ello contaremos con temperaturas de los dos días anteriores, históricos de máximas temperaturas y un amigo bastante chapas que te da su predicción del tiempo cada día.

**Resumen de datos**:
* year: 2016 for all data points
* month: number for month of the year
* day: number for day of the year
* week: day of the week as a character string
* temp_2: max temperature 2 days prior
* temp_1: max temperature 1 day prior
* average: historical average max temperature
* actual: max temperature measurement
* friend: your friend’s prediction, a random number between 20 below the average and 20 above the average


### 1. Importa el csv "temps.csv"

```python
import pandas as pd
features = pd.read_csv('data/temps.csv')
features.head()
```

```python
features.week.unique()
```

### 2. Lidia con las variables categóricas del dataset

```python
features = pd.get_dummies(features).drop(columns=['year'])
features.head()
```

### 3. Divide en train y test

```python
from sklearn.model_selection import train_test_split

y = features['actual']
X = features.drop('actual', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

### 4. Entrena el modelo
Utiliza un DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

### 5. Calcula su MAE
Sin usar sklearn

```python
import numpy as np
pred = dtr.predict(X_test)
error = abs(y_test - pred)
print('MAE:',np.round(np.mean(error),2),'degrees')
```

### 6. Calcula su MAPE
Sin usar sklearn

```python
error = abs(y_test - pred)
print('MAPE:',np.round(100*np.mean(error/y_test),2),'%')
```

### 7. Representa el árbol de decision

```python
import sklearn
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sklearn.tree.plot_tree(dtr, feature_names=X.columns,
                      class_names='actual',
                      filled=True);
```

### 8. Modifica el max_depth a 3 y vuelve a entrenarlo

```python
dtr3 = DecisionTreeRegressor(max_depth=3)
dtr3.fit(X_train, y_train)
```

### 9. Vuelve a calcular si MAE

```python
import numpy as np
pred = dtr3.predict(X_test)
error = abs(y_test - pred)
print('MAE:',np.round(np.mean(error),2),'degrees')
```

```python
error = abs(y_test - pred)
print('MAPE:',np.round(100*np.mean(error/y_test),2),'%')
```

### 10. Vuelve a representar su árbol

```python
import sklearn
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sklearn.tree.plot_tree(dtr3, feature_names=X.columns,
                      class_names='actual',
                      filled=True);
```

### 10. Obten el `feature_importances` de cada variable en el último modelo

```python
dtr3.feature_importances_
```

```python
pd.DataFrame({'features':X.columns,
              'importance':dtr3.feature_importances_}).sort_values('importance',
                                                                  ascending=False)
```
